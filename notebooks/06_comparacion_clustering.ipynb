# Ejercicio 2

En este segundo ejercicio se presenta otro dataset de juguete, donde hay tres clusters y cada uno de ellos tiene una densidad de puntos diferente al resto. El objetivo es probar las diferentes configuraciones de los cuatro algoritmos de clustering mencionados y analizar los resultados. Luego, responded a las siguientes preguntas:

* ¿Qué se esperaba de los algoritmos de clustering? ¿Cuántos clusters debería haber?

* ¿Qué algoritmo de clustering obtiene mejores resultados? ¿Por qué?

* ¿Crees que el índice de Silhouette es un buen indicador en este problema?

```python
sizes = np.array([10,100,1000])
avgs_x = np.array([1,4,10])
avgs_y = np.array([5,0,10])
```

```python
data3 = []

for a_x,a_y,s in zip(avgs_x,avgs_y,sizes):
    cx = []
    cy = []
    for  i in range(s):
        cx.append(random.gauss(a_x,1))
        cy.append(random.gauss(a_y,1))
    cx = np.array(cx)
    cy = np.array(cy)
    aux = np.vstack((cx,cy)).T
    data3.append(aux)

data = np.concatenate(data3, axis=0)
```

```python
plt.plot(data[:,0],data[:,1],'k.',alpha=0.2)
plt.show()
```
